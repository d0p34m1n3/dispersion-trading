```
import pandas as pd

index_info = pd.read_excel("/dbfs/FileStore/tables/Capstone.xlsx", sheetname=0)
index_info.shape
```

```
ii2 = index_info.set_index('Symbol')[['Index Weight','IV Index Call','IV Index Put']]
ii2.head(3)
```

```
s = ii2['Index Weight']
ii2['weight'] = s.str.strip('%').map(lambda x: float(x)/100)
ii2.loc['XOM']['IV Index Call']
```

```
import pandas as pd

correlation_matrix = pd.read_excel("/dbfs/FileStore/tables/Capstone.xlsx", 
                                   sheetname=1, skiprows=2)\
                       .rename(columns={'Unnamed: 1':'Symbol'})
cm2 = correlation_matrix.set_index('Symbol').drop('Unnamed: 0',axis=1).iloc[:31,]
```

```
cm2
```

```
from itertools import permutations, product, combinations_with_replacement
```

```
symbol_pairs = [(symbol_one, symbol_two) 
                for (symbol_one, symbol_two) 
                in  list(combinations_with_replacement(list(cm2.index),2))
                if symbol_one <= symbol_two
               ] 
```

```
import numpy as np
```

```
symbol_pair_correlation = [ np.prod((cm2[symbol_one][symbol_two],
                                     ii2.loc[symbol_one]['IV Index Call'],
                                     ii2.loc[symbol_two]['IV Index Call'],
                                     ii2.loc[symbol_one]['weight'],
                                     ii2.loc[symbol_two]['weight']))
 for (symbol_one, symbol_two)
 in symbol_pairs
]
symbol_pair_correlation
```

```
np.sqrt(sum(symbol_pair_correlation)*100**2)
```
